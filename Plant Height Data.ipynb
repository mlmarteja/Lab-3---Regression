```python
# Project by Marian Marteja and Andrew Kivrak
# Due: 11/05/21
##########################################################################################################
# Plant Height Data: In this project, we create scatterplots with the line of best fit of temperature,
#                    altitude, rain, and site number to plant height. In addtion, we calculated R sqaured
#                    for each data.
##########################################################################################################
# Methods:
# main(): This is the main function of the module. The main reads the csv of Plant height data using pandas.
#         The main will have variables that will store an aray of the the temp, height, rain, alt, and site
#         data. The program will print the R squared and display the scatter plot of each sets of data
# ScatterPlot(): gets the x and y values along with the title of the graph. In this method, the line of best
#                is made and the graphs are generated
#RSquared(): this method takes the x and y value and calculates the r-squared
##########################################################################################################

#imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



##Rain, Altitude, Site Number
def main():
    plantData = pd.read_csv("PlantHeightData.csv", 
                            usecols = ['temp','height','alt','rain','site'])
    ##x and y data sets
    temp = np.array(plantData["temp"])
    height = np.array(plantData["height"])
    rain = np.array(plantData["rain"])
    alt = np.array(plantData["alt"])
    site = np.array(plantData["site"])
    
    ##temp and height 
    print("R-Squared of temp and height: " + str(round(RSquared(temp,height),4)))
    ScatterPlot(temp,height,"Temperature and Height","temp","height")
    
    print("\n")
    
    ##rain and height
    print("R-Squared of rain and height: " + str(round(RSquared(rain,height),4)))
    ScatterPlot(rain,height,"Rain and Height","rain","height")
    
    print("\n")
    
   ##alt and height
    print("R-Squared of altitude and height: " + str(round(RSquared(alt,height),4)))
    ScatterPlot(alt,height,"Altitude and Height","alt","height")
    
    print("\n")
    
    ##site and height  
    print("R-Squared of site and height: " + str(RSquared(site,height)))
    ScatterPlot(site,height,"Site and Height","site","height")

##create scatter plot##
def ScatterPlot(data_x,data_y,title,xtitle,ytitle):
    plt.scatter(data_x,data_y)
    plt.title(title)
    plt.plot(data_x,data_y,'o')
    plt.xlabel(xtitle)
    plt.ylabel(ytitle)
    ##line of best fit
    m,b = np.polyfit(data_x,data_y,1)
    plt.plot(data_x, m*data_x + b)
    plt.show()


##r-squared##
def RSquared(xDataSet,yDataSet):
    correlation_matrix = np.corrcoef(xDataSet, yDataSet)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    return r_squared


##calling main##
main()
```
